```python
def combine_box(boxes_list):
    new_boxes_list = []
    for i in range(len(boxes_list)):
        current_box = boxes_list[i]
        merged = False

        for j in range(len(new_boxes_list)):
            new_box = new_boxes_list[j]
            overlap = calculate_overlap(current_box, new_box)
            
            if overlap:
                merged = True
                if current_box[7] > new_box[7]:
                    new_boxes_list[j] = current_box
                break

        if not merged:
            new_boxes_list.append(current_box)

    return new_boxes_list


def calculate_overlap(box1, box2):
    short_side_box1 = min(get_side_length(box1, 0), get_side_length(box1, 2))
    short_side_box2 = min(get_side_length(box2, 0), get_side_length(box2, 2))

    if short_side_box1 < short_side_box2:
        short_side = short_side_box1
    else:
        short_side = short_side_box2

    overlap_threshold = short_side * 0.2

    if abs(box1[0] - box2[0]) < overlap_threshold and abs(box1[1] - box2[1]) < overlap_threshold:
        return True

    return False


def get_side_length(box, index):
    x1, y1, x2, y2, x3, y3, x4, y4 = box
    side_length = ((box[index] - box[(index + 2) % 8]) ** 2 + (box[index + 1] - box[(index + 3) % 8]) ** 2) ** 0.5
    return side_length
```
